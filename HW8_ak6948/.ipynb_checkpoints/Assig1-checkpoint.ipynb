```python
import pylab as pl
import pandas as pd
import numpy as np
import os
import json

%pylab inline
```

```python
url = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD"
names = pd.read_csv(url)
names.head()
```

```python
names2016 = names[names["Year of Birth"]==2016]
names2016.head()
```

```python
malenames = names2016[names2016["Gender"]=="MALE"]
malenames.head()
```

```python
malenames.sort_values("Count")
topmalenames = malenames[0:10]
topmalenames
```

```python
femalenames = names2016[names2016["Gender"]=="FEMALE"]
femalenames.head()
```

```python
femalenames.sort_values("Count")
topfemalenames = femalenames[0:10]
topfemalenames
```

```python
pl.figure(figsize=(18,8))
pl.subplot(1, 2, 1)
bar(topmalenames["Child\'s First Name"],topmalenames.Count,)
pl.xticks(rotation=45, fontsize=12)
pl.yticks(fontsize=12)
pl.xlabel('Male Names', fontsize = 16) 
pl.ylabel('Count', fontsize = 16)

pl.subplot(1, 2, 2)
bar(topfemalenames["Child\'s First Name"],topfemalenames.Count,)
pl.xticks(rotation=45, fontsize=12)
pl.yticks(fontsize=12)
pl.xlabel('Female Names', fontsize = 16) 
pl.ylabel('Count', fontsize = 16)
pl.suptitle("Top 10 Male & Female Names in 2016", fontsize=14)

caption = """In this figure, the top 10 male and female names in 2016 are presented. 
         The top male name in 2016 was 'Liam', while the top female name was 'Isabella' """
pl.figtext(0.5, -0.15, caption, wrap=True, horizontalalignment='center', fontsize=20)
```
