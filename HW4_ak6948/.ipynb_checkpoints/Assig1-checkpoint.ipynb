```python
import numpy as np
import pylab as pl
d = {}
%pylab inline
```

```python
#==============================================================================
#CHI-SQUARED:
#==============================================================================
d['chisq'] = {}
df = 100
np.random.seed(1000)
mysize_chi = (np.random.rand(100) * 2000).astype(int) 
for n in mysize_chi:
    d['chisq'][n] = np.random.chisquare(df, size = n)

d['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in d['chisq']:
    if not type(nn) == str:
        d['chisq']['means'][nn] = d['chisq'][nn].mean()
        axchisq_mu_n.plot(nn, d['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize_chi), max(mysize_chi)], [df, df], 'k')

allmeans_chi = list(d['chisq']['means'].values())
pl.figure(figsize=(10, 10))
pl.hist(allmeans_chi,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

```python
#==============================================================================
#NORMAL DISTRIBUTION:
#==============================================================================
d['normal'] = {}
mean = 100
sd =  1
np.random.seed(999)
mysize_normal = (np.random.rand(100) * 2000).astype(int) 
for n in mysize_normal:
    d['normal'][n] = np.random.normal(mean, sd, size = n)

d['normal']['means'] = {}
axnormal_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in d['normal']:
    if not type(nn) == str:
        d['normal']['means'][nn] = np.mean(d['normal'][nn])
        axnormal_mu_n.plot(nn, d['normal']['means'][nn], 'o')
        axnormal_mu_n.set_xlabel('sample size', fontsize=18)
        axnormal_mu_n.set_ylabel('sample mean', fontsize=18)
        axnormal_mu_n.set_title('Normal Distribution', fontsize=18)
        axnormal_mu_n.plot([min(mysize_normal), max(mysize_normal)], [mean, mean], 'k')

allmeans_normal = list(d['normal']['means'].values())
pl.figure(figsize=(10, 10))
pl.hist(allmeans_normal,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

```python
#==============================================================================
#POISSON DISTRIBUTION:
#==============================================================================
d['poisson'] = {}
lambda_poisson = 2
np.random.seed(99)
mysize_poisson = (np.random.rand(100) * 2000).astype(int) 
for n in mysize_poisson:
    d['poisson'][n] = np.random.poisson(lambda_poisson, size = n)

d['poisson']['means'] = {}
axpoisson_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in d['poisson']:
    if not type(nn) == str:
        d['poisson']['means'][nn] = np.mean(d['poisson'][nn])
        axpoisson_mu_n.plot(nn, d['poisson']['means'][nn], 'o')
        axpoisson_mu_n.set_xlabel('sample size', fontsize=18)
        axpoisson_mu_n.set_ylabel('sample mean', fontsize=18)
        axpoisson_mu_n.set_title('Possion Distribution', fontsize=18)
        axpoisson_mu_n.plot([min(mysize_poisson), max(mysize_poisson)], [lambda_poisson, lambda_poisson], 'k')

allmeans_poisson = list(d['poisson']['means'].values())
pl.figure(figsize=(10, 10))
pl.hist(allmeans_poisson,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

```python
#==============================================================================
#BINOMIAL DISTRIBUTION:
#==============================================================================
d['binomial'] = {}
n_binomial = 10
p_binomial = 0.5
mean_binomial = n_binomial*p_binomial
np.random.seed(100)
mysize_binomial = (np.random.rand(100) * 2000).astype(int) 
for n in mysize_binomial:
    d['binomial'][n] = np.random.binomial(n_binomial, p_binomial, size = n)

d['binomial']['means'] = {}
axbinomial_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in d['binomial']:
    if not type(nn) == str:
        d['binomial']['means'][nn] = np.mean(d['binomial'][nn])
        axbinomial_mu_n.plot(nn, d['binomial']['means'][nn], 'o')
        axbinomial_mu_n.set_xlabel('sample size', fontsize=18)
        axbinomial_mu_n.set_ylabel('sample mean', fontsize=18)
        axbinomial_mu_n.set_title('Binomial Distribution', fontsize=18)
        axbinomial_mu_n.plot([min(mysize_binomial), max(mysize_binomial)], [mean_binomial, mean_binomial], 'k')

allmeans_binomial = list(d['binomial']['means'].values())
pl.figure(figsize=(10, 10))
pl.hist(allmeans_binomial,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

```python
#==============================================================================
#LAPLACE DISTRIBUTION:
#==============================================================================
d['laplace'] = {}
mu = 100
b_laplace = 1
np.random.seed(10)
mysize_laplace = (np.random.rand(100) * 2000).astype(int) 
for n in mysize_laplace:
    d['laplace'][n] = np.random.laplace(mu, b_laplace, size = n)

d['laplace']['means'] = {}
axlaplace_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in d['laplace']:
    if not type(nn) == str:
        d['laplace']['means'][nn] = np.mean(d['laplace'][nn])
        axlaplace_mu_n.plot(nn, d['laplace']['means'][nn], 'o')
        axlaplace_mu_n.set_xlabel('sample size', fontsize=18)
        axlaplace_mu_n.set_ylabel('sample mean', fontsize=18)
        axlaplace_mu_n.set_title('Laplace Distribution', fontsize=18)
        axlaplace_mu_n.plot([min(mysize_laplace), max(mysize_laplace)], [mu, mu], 'k')

allmeans_laplace = list(d['laplace']['means'].values())
pl.figure(figsize=(10, 10))
pl.hist(allmeans_laplace,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```
